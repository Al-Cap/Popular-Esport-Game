```python
## using xlrd to read excel files
## using pandas to import and show data in table formate
import matplotlib.pyplot as plt
import numpy as np
import xlrd
import pandas as pd

## data is imported from an excel file
data = pd.read_excel("../data/EsportCorrected_1.xlsx")
data
```

```python
# compiling data to show the limitatins of the data
data.describe()
```

```python
# show data types used
list(set(data.dtypes))
```

```python
# list data columns
data.columns.to_list()
```

```python
# list Game column's index
data.Game.to_list()
```

```python
# list Genre column's index
data.Genre.to_list()
```

```python
# list Release Date column's index
data.ReleaseDate.to_list()
```

```python
# list Total Earnings column's index
data.TotalEarnings.to_list()
```

```python
# set variables for earnings and genre
total_earnings = data.TotalEarnings.to_list()
genre = data.Genre.to_list()

# graphing in horizontal bar graph of genre and earngings
plt.barh(genre, total_earnings)
plt.xlabel('Total Earnings [milions]')
plt.ylabel('Genre')
plt.title('Earnings Per Genre')
plt.savefig('../figures/genre-tearn_barh.png')
```

```python
# set variables for release date and tournaments
rel = data.ReleaseDate.to_list()
tour = data.TotalTournaments.to_list()

# graphing in bar graph of release date and tournaments
plt.bar(rel, tour)
plt.xlabel('Release Date of Games')
plt.ylabel('Total Tournaments')
plt.title('Tournaments Per Year')
plt.savefig('../figures/tour-rel_bar.png')
```

```python
# set variables for total players
ply = data.TotalPlayers.to_list()

# graphing in horizontal bar graph of total players and genre
plt.barh(genre, ply)
plt.xlabel('Total Players')
plt.ylabel('Genre')
plt.title('Players Per Genre')
plt.savefig('../figures/genre-ply_barh.png')
```

```python
# compressed data to view only first-person shooter games
# assigned to a new variable
ds = data[data.Genre == 'First-Person Shooter']
ds = ds[ds.TotalPlayers > 1000]
ds
```

```python
# set variable for gaame under the new data set
# redefined total players under new data set
gs = ds.Game.to_list()
ps = ds.TotalPlayers.to_list()

# graphing in horizontal bar graph of players and games
plt.barh(gs, ps)
plt.xlabel('Players')
plt.ylabel('Game')
plt.title('Players Per Game')
plt.savefig('../figures/FirstP-Shooter_game-player_barh')
```

```python
# graphing in horizontal bar graph of genres and tournaments
plt.barh(genre, tour)
plt.ylabel('Genre')
plt.xlabel('Total Tournaments')
plt.title('Tournaments Per Genres')
plt.savefig('../figures/genre-tour_barh.png')
```
