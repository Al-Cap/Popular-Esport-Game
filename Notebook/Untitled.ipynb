```python
## using xlrd to read excel files
## using pandas to import and show data in table formate
import matplotlib.pyplot as plt
import numpy as np
import xlrd
import pandas as pd

## data is imported from an excel file
data = pd.read_excel("../data/EsportCorrected_1.xlsx")
data
```

```python
total_earnings = data.TotalEarnings.to_list()
genre = data.Genre.to_list()

plt.barh(genre, total_earnings)
plt.xlabel('Total Earnings [milions]')
plt.ylabel('Genre')
plt.title('The Total Earnings in Genre')
plt.savefig('../figures/genre-tearn_barh.png')
```

```python
rel = data.ReleaseDate.to_list()
tour = data.TotalTournaments.to_list()

plt.bar(rel, tour)
plt.xlabel('Release Date of Games')
plt.ylabel('Total Tournaments')
plt.title('The Amount of Tournaments Ocurring Yearly')
plt.savefig('../figures/tour-rel_bar.png')
```

```python
ply = data.TotalPlayers.to_list()

plt.barh(genre, ply)
plt.xlabel('Total Players')
plt.ylabel('Genre')
plt.title('Total Players Per Genre')
plt.savefig('../figures/genre-ply_barh.png')
```

```python
ds = data[data.Genre == 'First-Person Shooter']
ds = ds[ds.TotalPlayers > 1000]
ds
```

```python
gs = ds.Game.to_list()
ps = ds.TotalPlayers.to_list()

plt.barh(gs, ps)
plt.xlabel('Players')
plt.ylabel('Game')
plt.title('Number of Players Per Game')
plt.savefig('../figures/FirstP-Shooter_game-player_barh')
```
